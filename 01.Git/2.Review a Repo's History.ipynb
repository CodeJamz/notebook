## Git Log
The `git log` command is used to display all of the commits of a repository.<br>
*`git log`命令用于显示存储库的所有提交。*
```
$ git log
```

By default, this command displays(*默认情况下，此命令显示*):
- the SHA(*SHA*)
- the author(*作者*)
- the date(*日期*)
- the message(*消息*)

…of every commit in the repository(*…存储库中的每个提交。*)

Git uses the command line pager, Less, to page through all of the information. The important keys for Less are:<br>
*Git使用命令行分页器Less来分页所有信息。 Less的重要键是：*
- to scroll down by a line, use `j` or `↓`(*向下滚动一行，使用 `j`或 `↓`*)
- to scroll up by a line, use `k` or `↑`(*要向上滚动一行，请使用 `k`或 `↑`*)
- to scroll down by a page, use the `spacebar` or the `Page Down` button(*要向下滚动页面，请使用 `空格键`或 `Page Down`按钮*)
- to scroll up by a page, use `b` or the `Page Up` button(*要向上滚动一页，请使用 `b`或 `Page Up`按钮*)
- to quit, use `q`(*退出，用 `q`*)


## git log --oneline
the `--oneline` flag is used to alter how `git log` displays information:<br>
*`--oneline`标志用于更改 `git log`显示信息的方式：*
```
$ git log --oneline
```

This command:
- lists one commit per line(*每行列出一次提交*)
- shows the first 7 characters of the commit's SHA(*显示提交的SHA的前7个字符*)
- shows the commit's message(*显示提交的消息*)

## git log --stat
the `--stat` flag is used to alter how `git log` displays information:<br>
* `--stat`标志用于更改`git log`显示信息的方式：
```
$ git log --stat
```
This command:
- displays the file(s) that have been modified(*显示已修改的文件*)
- displays the number of lines that have been added/removed(*显示已添加/删除的行数*)
- displays a summary line with the total number of modified files and lines that have been added/removed<br>
(*显示摘要行，其中包含已修改文件的总数以及已添加/删除的行*)

## git log -p
the `-p` flag (which is the same as the `--patch` flag) is used to alter how `git log` displays information:<br>
*`-p`标志（与`--patch`标志相同）用于更改 `git log`显示信息的方式：*
```
$ git log -p
```

This command adds the following to the default output:<br>
*此命令将以下内容添加到默认输出：*
- displays the files that have been modified(*显示已修改的文件*)
- displays the location of the lines that have been added/removed(*显示已添加/删除的行的位置*)
- displays the actual changes that have been made(*显示已进行的实际更改*)

**Annotated `git log -p` Output**<br>
***带注释的 `git log -p`输出***<br>

- 🔵 - the file that is being displayed<br>
 (*正在显示的文件*)
 
 
- 🔶 - the hash of the first version of the file and the hash of the second version of the file<br>
 (*文件第一个版本的哈希值和文件第二个版本的哈希值*)
 - not usually important, so it's safe to ignore<br>
  (*通常不重要，因此可以放心忽略*)
  
  
- ❤️ - the old version and current version of the file<br>
 (*文件的旧版本和当前版本*)
 
 
- 🔍 - the lines where the file is added and how many lines there are<br>
 (*添加文件的行以及多少行*)
 - `-15,83` indicates that the old version (represented by the `-`) started at line 15 and that the file had 83 lines<br>
  (*`-15,83`表示旧版本（由 `-`表示）从第15行开始，并且文件有83行*)
 - `+15,85` indicates that the current version (represented by the `+`) starts at line 15 and that there are now 85 lines…these 85 lines are shown in the patch below<br>
  (* `+15,85`表示当前版本（用 `+`表示）从第15行开始，现在有85行…这些85行显示在下面的补丁中*)
  
  
- ✏️ - the actual changes made in the commit<br>
 (*提交中的实际更改*)
 - lines that are red and start with a minus (`-`) were in the original version of the file but have been removed by the commit<br>
  (*红色且以减号（`-`）开头的行在文件的原始版本中，但是已被提交删除*)
 - lines that are green and start with a plus (`+`) are new lines that have been added in the commit<br>
  (*绿色且以加号（`+`）开头的行是已在提交中添加的新行*)

## Git Show
The `git show` command will show only one commit.The output of the `git show` command is exactly the same as the `git log -p` command. <br>
(*`git show`命令将仅显示一次提交。 因此，当您找不到任何其他提交时，请不要惊慌-它只会显示一个。 `git show`命令的输出与`git log -p`命令完全相同。<br>

Running it like the example will only display the most recent commit.<br>
(*这样样运行它只会显示最新的提交。*)
```
$ git show
```

Typically, a SHA is provided as a final argument:<br>
(*通常，将SHA作为最终参数提供*)
```
$ git show fdf5493
```
by default, `git show` displays:
- the commit(*提交*)
- the author(*作者*)
- the date(*日期*)
- the commit message(*提交消息*)
- the patch information(*补丁信息*)

However, `git show` can be combined with most of the other flags we've looked at:<br>
(*然而，`git show`可以与我们查看过的大多数其他标志结合使用*)

- `--stat` - to show the how many files were changed and the number of lines that were added/removed<br>
 `--stat`-显示更改了多少文件以及添加/删除的行数
- `-p` or `--patch` - this the default, but if `--stat` is used, the patch won't display, so pass `-p` to add it again<br>
 `-p`或`--patch`-这是默认设置，但是如果使用`--stat`，则补丁不会显示，因此请传递`-p`再次添加
- `-w` - to ignore changes to whitespace<br>
 `-w`-忽略对空格的更改
