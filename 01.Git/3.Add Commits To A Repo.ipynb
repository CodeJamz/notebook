## Git Add
The `git add` command is used to move files from the Working Directory to the Staging Index.<br>
*`git add`命令用于将文件从工作目录移动到暂存索引*)

```
$ git add <file1> <file2> … <fileN>
```
command:
- takes a space-separated list of file names<br>
 *以空格分隔的文件名列表*
- alternatively, the period `.` can be used in place of a list of files to tell Git to add the current directory (and all nested files)<br>
 *或者 `.`可以代替文件列表来告诉Git添加当前目录（以及所有嵌套文件）*

## Git Commit
The `git commit` command takes files from the Staging Index and saves them in the repository.<br>
*`git commit`命令从暂存索引中获取文件，并将其保存在存储库中。*
```
$ git commit
```

This command:
- will open the code editor that is specified in your configuration(*将打开您的配置中指定的代码编辑器*)

Inside the code editor:
- a commit message must be supplied (*必须提供提交消息*)
- lines that start with a # are comments and will not be recorded (*以＃开头的行是注释，不会被记录*)
- save the file after adding a commit message (*添加提交消息后保存文件*)
- close the editor to make the commit (*关闭编辑器进行提交*)

**TIP：**If the commit message you're writing is short and you don't want to wait for your code editor to open up to type it out, you can pass your message directly on the command line with the -m flag:<br>
***提示：*** *如果正在编写的提交消息很短，并且不想等待代码编辑器打开以键入它，则可以使用-m标志直接在命令行中传递消息：*
```
$ git commit -m "Initial commit"
```
In the example above, the text `Initial commit` is used as the commit message. Be aware that you can't provide a description for the commit, only the message part.<br>
*在上面的示例中，文本 `Initial commit`被用作提交消息。 请注意，您无法提供有关提交的描述，只能提供消息部分,也就是第一行*

Only the message (the first line) is included in `git log --oneline`<br>
*只有消息（第一行）包含在 `git log --oneline`中！*

## Git Diff
The `git diff` command can be used to see changes that have been made but haven't been committed, yet<br>
*`git diff`命令用于查看已进行但尚未提交的更改*
```
$ git diff
```

This command displays:
- the files that have been modified (*已修改的文件*)
- the location of the lines that have been added/removed (*已添加/删除的行的位置*)
- the actual changes that have been made (*实际进行的更改*)


## Git Ignore
the `.gitignore` file is used to tell Git about the files that Git should not track. This file should be placed in the same directory that the `.git` directory is in.<br>
*`.gitignore`文件用于告知Git不应跟踪的文件。 该文件应放在`.git`目录所在的目录中*

Globbing lets you use special characters to match patterns/characters. In the `.gitignore` file, you can use the following:<br>
*通过globbing，您可以使用特殊字符来匹配模式/字符。 在 `.gitignore`文件中，可以使用以下命令：*
- blank lines can be used for spacing (*空行可用于间隔*)
- #- marks line as a comment (*＃将行标记为注释*)
- ? - matches 1 character (*？匹配1个字符*)
- [abc] - matches a, b, or c (*[abc]匹配a，b或c*)
- ** - matches nested directories - a/**/z matches (* *匹配嵌套目录*)
 - a/z
 - a/b/z
 - a/b/c/z

So if all of the 50 images are JPEG images in the "samples" folder, we could add the following line to .gitignore to have Git ignore all 50 images.<br>
因此，如果“ samples”文件夹中的所有50张图像都是JPEG图像，我们可以在.gitignore中添加以下行，以使Git忽略所有50张图像。
```
samples/*.jpg
```
