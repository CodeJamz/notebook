## Create a Pull Request
A **pull request** is a request for the source repository to pull in your commits and merge them with their project. To create a pull request, a couple of things need to happen:<br>
**提取请求**是对源存储库的请求，以提取您的提交并将其与项目合并。 要创建请求请求，需要完成以下几件事：<br>
- you must **fork** the source repository (*您必须fork源存储库*)
- clone your fork down to your machine (*将fork克隆到计算机上*)
- make some commits (ideally on a topic branch) (*进行一些提交,最好在主题分支上*)
- push the commits back to your fork (*将提交推回fork*)
- create a new pull request and choose the branch that has your new commits (*创建一个新的拉取请求并选择具有新提交的分支*)

## Stay in sync with source project
While you're working on a topic branch of changes that you want to make to a repository, that repository will probably be receiving updates of its own from the original authors.<br>
*当您要对存储库进行更改的主题分支时，该存储库可能会收到来自原始作者的更新。*<br>

**Stars & Watching**<br>
If you want to keep up-to-date with the Repository, GitHub offers a convenient way to keep track of repositories - it lets you star repositories<br>
*如果您想了解存储库的最新信息，GitHub提供了一种方便的方式来跟踪存储库-它使您可以对存储库加注星标*<br>
Starring is helpful if you want to keep track of certain repositories. But it's not entirely helpful if you need to actively keep up with a repositories development because you have to manually go to the stars page to view the repositories and see if they've changed.<br>
*如果您想跟踪某些存储库，加星号将很有帮助。 但是，如果您需要积极地跟上存储库的开发，这并不是完全有帮助的，因为您必须手动转到stars页面以查看存储库并查看它们是否已更改。*<br>

**Watching A Repository**<br>
If you need to keep up with a project's changes and want to be notified of when things change, GitHub offers a "Watch" feature<br>
*如果您需要跟上项目的更改并希望在发生更改时收到通知，则GitHub提供了“监视”功能<br>*
The Notification settings. "Watching" a repository will alert you to all activity.<br>
*通知设置。 “监视”存储库将提醒您所有活动。*<br>
If you're working on a repository quite often, then I'd suggest setting the watch setting to "Watching". This way GitHub will notify you whenever anything happens with the repository like people pushing changes to the repository, new issues being created, or comments being added to existing issues.<br>
*如果您经常在存储库上工作，则建议将监视设置设置为“监视”。 这样，GitHub会在存储库发生任何情况时通知您，例如人们将更改推送到存储库，创建新问题或将注释添加到现有问题。*<br>

**Including Upstream Changes**<br>
Let's see how we can get these changes from the remote repository into our repository.<br>
*让我们看看如何将这些更改从远程存储库获取到我们的存储库中。*<br>

In my local repository, I already have one remote repository which is `origin` remote.<br>
*在我的本地存储库中，我已经有一个远程存储库，它是 `origin`远程存储库。*<br>
Remember that the word `origin` is just the default name that's used when you `git clone` a remote repository for the first time.<br>
*请记住，单词 `origin`只是您第一次 `git clone`远程存储库时使用的默认名称。* <br>
We're going to use the `git remote` command to add a new shortname and URL to this list. This will give us a connection to the source repository.<br>
*我们将使用 `git remote`命令向该列表添加新的短名称和URL。 这将使我们与源存储库建立连接。* <br>
```
$ git remote add upstream https://github.com/udacity/course-collaboration-travel-plans.git
```
Notice that I've used the name `upstream` as the shortname to reference the source repository.<br>
*请注意，我已经使用`upstream`作为简称来引用源存储库。*<br>
Let's check out what the list of remotes looks like now after adding this new remote:<br>
*在添加此新遥控器之后，让我们检查一下遥控器的列表现在是什么样的：*<br>
```
$ git remote -v
```
The terminal application show both information about both remotes - `origin` and `upstream`.<br>
*终端应用程序同时显示有关远程服务器的信息-`origin`和`upstream`。*<br>

**Retrieving Upstream Changes**<br>
Now to get the changes from upstream remote repository, all we have to do is run a `git fetch` and use the `upstream` shortname rather than the `origin` shortname:<br>
*现在要从上游远程存储库获取更改，我们要做的就是运行 `git fetch`并使用 `upstream`短名称而不是 `origin`短名称：*<br>
```
$ git fetch upstream master
```
The terminal application showing the results of doing `git fetch upstream master`. A new branch is added to the local repository.<br>
*终端应用程序显示执行`git fetch upstream master`的结果。 新分支将添加到本地存储库。*<br>

let's do a log to see what new information we have in our local repository. I'm using the following `git log` command to make sure I display all commits from all branches (including remote and tracking branches!):<br>
*让我们做一个日志，看看我们本地存储库中有什么新信息。 我正在使用以下 `git log`命令来确保显示所有分支（包括远程和跟踪分支！）的所有提交：*<br>
```
$ git log --oneline --graph --decorate --all
```
The terminal application showing the log of my local repository after fetching the `upstream` remote's changes.<br>
终端应用程序在获取 `upstream`远程服务器的更改后显示我的本地存储库的日志。<br>
We can use the `upstream/master` branch to keep track of where the source repository's master branch is. We can now get any changes that are made to the source repository's `master` branch by just running `git fetch upstream master`.<br>
我们可以使用 `upstream/master`分支来跟踪源存储库的主分支在哪里。 现在，我们只需运行 `git fetch upstream master`，就可以对源存储库的 `master`分支进行任何更改。<br>

we don't want to run `git push origin upstream/master` because `upstream/master` is not a local branch.<br>
我们不想在`upstream/master`服务器上运行git push origin，因为`upstream/master`服务器不是本地分支。<br>
To get these changes into my forked version of her project, I could merge `upstream/master` into an existing branch (like the local `master` branch) and push that.<br>
为了将这些更改纳入我的项目的分支版本中，我可以将 `upstream/master`版合并到现有分支（如本地 `master`分支）中并进行推送。
```
# to make sure I'm on the correct branch for merging (*确保我在正确的分支上进行合并*)
$ git checkout master

# merge in Lam's changes (*合并林的变化*)
$ git merge upstream/master

# send owner's changes to my remote (*将主人的更改发送到我的远程*)
$ git push origin master
```

**Recap**<br>
When working with a project that you've forked. The original project's maintainer will continue adding changes to their project. You'll want to keep your fork of their project in sync with theirs so that you can include any changes they make.<br>
*在处理您分叉的项目时。 原始项目的维护者将继续在其项目中添加更改。 您需要使他们的项目分支与他们的项目保持同步，以便您可以包括他们所做的任何更改。*<br>

To get commits from a source repository into your forked repository on GitHub you need to:<br>
*要将提交从源存储库提交到GitHub上的分叉存储库，您需要：*<br>
- get the cloneable URL of the source repository 
 *获取源存储库的可克隆URL*
- create a new remote with the `git remote add` command 
 *使用 `git remote add`命令创建一个新的遥控器*
 - use the shortname `upstream` to point to the source repository 
  *使用upstream缩写名称指向源存储库*
 - provide the URL of the source repository 
  *提供源存储库的URL*
- fetch the new `upstream` remote 
 *获取新的 `upstream`遥控器*
- merge the `upstream`'s branch into a local branch 
 *合并 `upstream`分支到本地分支*
- push the newly updated local branch to your `origin` repo 
 *将新近更新的本地分支推送到您的 `origin`仓库*

## Manage an active PR
The project maintainer may decide not to accept your changes right away. They might request you to make some additional changes to your code before accepting your request and merging in your changes. Most likely they will communicate their desired changes through the conversation on the pull requests page.<br>
项目维护者可能会决定不立即接受您的更改。 他们可能会要求您在接受您的请求并合并您的更改之前对代码进行一些其他更改。 他们很可能会通过拉取请求页面上的对话来传达所需的更改。<br>

If the project's maintainer is requesting changes to the pull request, then:<br>
*如果项目的维护者正在请求更改拉取请求，则：*
- make any necessary commits on the same branch in your local repository that your pull request is based on<br>
 *在您的拉取请求基于的本地存储库的同一分支上进行任何必要的提交*
- push the branch to the your fork of the source repository<br>
 *将分支推送到源存储库的分支*
The commits will then show up on the pull request page.<br>
*然后，提交将显示在拉取请求页面上。*

## Squash Commits
The `git rebase` command is used to do a great many things.
```
# interactive rebase
$ git rebase -i <base>

# interactively rebase the commits to the one that's 3 before the one we're on 
# 交互式地将提交的基准重新设置为之前3个的提交
$ git rebase -i HEAD~3
```
Inside the interactive list of commits, all commits start out as `pick`, but you can swap that out with one of the other commands<br>
*在交互式提交列表中，所有提交都是从`pick`开始的，但是您可以将其替换为其他命令之一*<br>
I recommend that you create a backup branch before rebasing, so that it's easy to return to your previous state. If you're happy with the rebase, then you can just delete the backup branch!<br>
*我建议您在重新设置基础之前创建一个备份分支，以便轻松返回到先前的状态。 如果您对重新设置感到满意，则可以删除备份分支！*

**Rebase Commands**<br>
Let's take another look at the different commands that you can do with `git rebase`:
- use `p` or `pick` – to keep the commit as is (*将提交保持原样*)
- use `r` or `reword` – to keep the commit's content but alter the commit message (*保留提交内容，但更改提交消息*)
- use `e` or `edit` – to keep the commit's content but stop before committing so that you can: (*保留提交的内容，但是在提交之前停止，这样您可以：*)
 - add new content or files (*添加新的内容或文件*)
 - remove content or files (*删除内容或文件*)
 - alter the content that was going to be committed (*更改将要提交的内容*)
- use `s` or `squash` – to combine this commit's changes into the previous commit (the commit above it in the list) (*将此提交的更改合并到上一个提交,列表中位于其上方的提交*)
- use `f` or `fixup` – to combine this commit's change into the previous one but drop the commit message (*将此次提交的更改合并到上一次更改中，但删除提交消息*)
- use `x` or `exec` – to run a shell command (*运行shell命令*)
- use `d` or `drop` – to delete the commit (*删除提交*)


## Force Pushing
I used `git rebase` to combine or squash all of these commits into one, single commit.<br>
*我使用 `git rebase`将所有这些提交合并或压缩为一个提交。*<br>
Using git rebase creates a new commit with a new SHA. When I tried using git push to send this commit up to GitHub, GitHub knew that accepting the push would erase the separate commits, so it rejected it.<br>
*使用git rebase创建具有新SHA的新提交。 当我尝试使用git push将提交提交到GitHub时，GitHub知道接受推送将删除单独的提交，因此它拒绝了。*<br>
So I had to force push the commits through using `git push -f`.<br>
*所以我不得不使用 `git push -f`强制推送提交。*
```
$ git push -f
```
