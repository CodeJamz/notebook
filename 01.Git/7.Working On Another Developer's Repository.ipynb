## Forking A Repository
if a repository doesn't belong to your account then it means you do not have permission to modify it.<br>
*如果存储库不属于您的帐户，则意味着您无权对其进行修改*<br>
Instead of modifying the original repository directly, if you fork the repository to your own account then you will have full control over that repository.<br>
*如果您将存储库分叉到自己的帐户中，则可以直接控制该存储库，而不是直接修改原始存储库*<br>
**Push/Pull To The Fork**<br>
It'll be pushing the changes back to your remote repository not to the original remote repository that you forked from.<br>
*它将把更改推回到您的远程存储库，而不是推回到您从中派生的原始远程存储库。*

## Reviewing Existing Work
We can discover details about what other developers have done by using the extremely powerful git log command.<br>
*我们可以使用功能非常强大的git log命令发现有关其他开发人员所做工作的详细信息。*<br>

**Group By Commit Author 分组作者**<br>
A quick way that we can see how many commits each contributor has added to the repository is to use the `git shortlog` command:<br>
*我们可以使用 `git shortlog`命令来快速查看每个贡献者已添加到存储库中的提交数量：*
```
 $ git shortlog
```
> It displays all commits sorted by author.<br>
> *它显示按作者排序的所有提交。*<br>

If we just want to see just the number of commits that each developer has made, we can add a couple of flags:<br>
*如果我们只想查看每个开发人员的提交次数，则可以添加几个标志：*
- `-s` to show just the number of commits (rather than each commit's message) (*`-s`仅显示提交的数量（而不是每个提交的消息）*)
- `-n` to sort them numerically (rather than alphabetically by author name) (*`-n`显示它们的编号（而不是按作者姓名的字母顺序）*)

```
$ git shortlog -s -n

```

> `git shortlog` with the `-s -n` flags to show only the number of commits each author has made, sorted numerically.<br>
> *带有 `-s -n`标志的 `git shortlog`，仅显示每个作者进行的提交数量，按数字排序。*<br>

**Filter By Author 按作者筛选**

Another way that we can display all of the commits by an author is to use the regular `git log` command but include the `--author` flag to filter the commits to the provided author.<br>
*我们可以显示作者的所有提交的另一种方法是使用常规的 `git log`命令，但包括 `--author`标志以过滤对提供的作者的提交*<br>
```
$ git log --author=name
```
> The output displays only the commits that Surma made.<br>
> *输出仅显示name所做的提交。*

**Filter Commits By Search 按搜索过滤提交**

If you write a descriptive commit message, then it's so much easier to search through the commit messages, later, to find exactly what you're looking for.<br>
*如果您编写了描述性的提交消息，那么以后搜索这些提交消息会变得更加容易，以至于可以准确地找到您要查找的内容。*<br>
e.g:Let see an example of extra details in a commit in the project by looking at commit `5966b66`:<br>
*通过查看commit `5966b66`，来看看项目中一个commit的其他细节示例：*
```
$ git show 5966b66
```
> show information about the commit `5966b66`.

We can filter commits with the `--grep` flag.<br>
*我们可以使用 `--grep`标志过滤提交*<br>
How about we filter down to just the commits that reference the word "bug". We can do that with either of the following commands:<br>
*我们如何过滤掉仅引用单词“ bug”的提交。 我们可以使用以下任一命令来实现：*
```
$ git log --grep=bug
$ git log --grep bug
```

* Watch Out For Spacing 注意间距<br>
 Remember that spacing is an issue, here, too. If you're trying to search for something that is multiple words and has spaces between the words, you need to wrap everything in quotes.<br>
 *请记住，间距也是一个问题。 如果您要搜索多个单词并且单词之间有空格的内容，则需要将所有内容都用引号引起来。*<br>
 For example, to search for `unit tests` you would need to use the following command, `git log --grep="unit tests"`.<br>
 *例如，要搜索 `unit tests`，您将需要使用以下命令，`git log --grep =“ unit tests”`*

## CONTRIBUTING.md File
The name of the `CONTRIBUTING.md` file is typically written in all caps so that it's easily seen. <br>
*通常将 `CONTRIBUTING.md`文件的名称写成全大写，以便于查看。* <br>
As you could probably tell by its name, this file lists out the information you should follow to contribute to the project. <br>
*您可能会用它的名字来告诉它，该文件列出了为该项目做贡献应遵循的信息。* <br>
You should look for this file before you start doing development work of any kind.  <br>
*在开始进行任何形式的开发工作之前，应先查找此文件。* <br>

There are two main sections to this file:
- the "For Contributors" section (*“供稿者”部分*)
- the "For Maintainers" section (*“维护者”部分*)

Each one of these sections has subsections of its own to instruct readers on how to contribute to and work with this project.  <br>
*这些部分中的每个部分都有其自己的子部分，以指导读者如何为该项目做出贡献并与之合作。* <br>

But if your change is more substantial where it modifies a number of files in a significant way, then you probably want to get approval by the project's maintainer(s) before you start working on it. You definitely don't want to spend a couple hours making changes to the project only to find out that someone else is doing the exact same thing. You'd just be wasting a lot of time and energy duplicating work.<br>
*如果您所做的更改更为重大，可以显着地修改多个文件，那么您可能希望在开始工作之前获得项目维护者的批准。 您绝对不想花几个小时对项目进行更改，只是为了发现其他人正在做的事情完全相同。 您只是在浪费大量时间和精力进行重复工作。* <br>
GitHub has a fantastic interface for asking questions of the project maintainer in an open way that lets everyone see what's being done with the project. <br>
*GitHub有一个很棒的界面，可以以开放的方式询问项目维护者的问题，让每个人都可以看到该项目正在做什么。* <br>

**issue tractor** <br>
Now, "issues" doesn't mean that there's actually a bug, it can just be any change that needs to be made to the project.  <br>
*现在，“问题”并不意味着实际上存在一个错误，它可以是需要对该项目进行的任何更改。* <br>

Each issue can:
- have a label or multiple labels applied to it (*带有一个或多个标签*)
- can be assigned to an individual (*可以分配给个人*)
- can be assigned a milestone (for example the issue will be resolved by the next major release) (*可以分配一个里程碑（例如，该问题将在下一个主要版本中解决）*)  <br>

But probably one of the most important aspects of the issue tracker is that each issue can have its own comments, so a conversation can form around the issue. <br>
*但是问题跟踪器最重要的方面之一可能是每个问题都可以有自己的评论，因此可以围绕该问题进行对话。* <br>

Another thing that's nice about issues is:
- they let you subscribe to an issue so you'll be notified of new comments and code changes <br>
 *他们让您订阅问题，以便您收到新评论和代码更改的通知* <br>
- you can communicate back and forth with a project maintainer on a specific change <br>
 *您可以就特定更改与项目维护者来回交流* <br>

- Before you contribute anything to a file, check out the instructions in `CONTRIBUTING.md`. Then check out the project's issues and look to see if there's anything that's similar to what you want to contribute. If there is, then subscribe to that issue and read the existing conversation to see if you can help. <br>
 *在为文件添加任何内容之前，请查看 `CONTRIBUTING.md`中的说明。 然后检查项目的问题，看看是否有任何与您想要贡献的东西相似的东西。 如果有，请订阅该问题并阅读现有对话以查看是否可以提供帮助。* <br>
- If you've looked through the list of issues and don't see one that similar to what you want to do, then you can create a new issue of your own. <br>
 *如果您浏览了问题列表，但未发现与您要执行的操作类似的问题，则可以创建自己的新问题。* <br>
 
**New Issue Page** <br>
The GitHub issues interface support markdown so when you create your issue you can use Markdown to format it and exactly the way you want by including links, images, bulleted lists, and code blocks. <br> 
*GitHub的问题界面支持 markdown，因此当您创建问题时，您可以使用Markdown对其进行格式化，并通过添加链接，图像，项目符号列表和代码块来完全按照所需的方式进行格式化。*

## Topic Branches 
The best way to organize the set of commits/changes you want to contribute back to the project is to put them all on a **topic branch**.<br>
*整理要向项目贡献的一组提交/更改的最好方法是将它们全部放在一个**主题分支**上。*<br>
Unlike the master branch which is the default branch that holds all of the commits for your entire project, a topic branch host commits for just a single concept or single area of change.<br>
*与master分支不同，master分支是保存整个项目的所有提交的默认分支，而topic分支主持人仅针对单个概念或单个更改区域提交。*<br>
There are plenty of names that can be used for a topic branch's name. You just want to use a clear descriptive name for the branch so that if, for example, you list out all of the branches you can immediately see what changes *are supposed to be in a branch just by its name.<br>
有很多名称可用于主题分支的名称。 您只想为分支使用一个明确的描述性名称，这样，例如，如果您列出了所有分支，则可以立即查看仅凭其名称应该在分支中进行的更改。*<br>
- definitely check out the CONTRIBUTING.md file to see if they provide instructions on what you should name your topic branches.<br>
 *一定要检查CONTRIBUTING.md文件，以查看它们是否提供有关应为主题分支命名的说明。*

## Recap
Before you start doing any work, make sure to look for the project's CONTRIBUTING.md file.<br>
*在开始任何工作之前，请确保查找项目的CONTRIBUTING.md文件。*<br>

Next, it's a good idea to look at the GitHub issues for the project<br>
*接下来，最好查看该项目的GitHub问题*<br>
- look at the existing issues to see if one is similar to the change you want to contribute<br>
 *查看现有问题，看看是否与您要贡献的变更相似*<br>
- if necessary create a new issue<br>
 *如有必要，创建一个新的问题*<br>
- communicate the changes you'd like to make to the project maintainer in the issue<br>
 *在问题中向项目维护者传达您要进行的更改*<br>

When you start developing, commit all of your work on a topic branch:<br>
*开始开发时，将所有工作提交到主题分支：*<br>
- do not work on the master branch<br>
 *在master分支上不起作用*<br>
- make sure to give the topic branch clear, descriptive name<br>
 *确保给主题分支提供清晰的描述性名称*<br>

As a general best practice for writing commits:<br>
*作为编写提交的一般最佳实践：*<br>
- make frequent, smaller commits<br>
 *频繁进行较小的提交*<br>
- use clear and descriptive commit messages<br>
 *使用清晰和描述性的提交消息*<br>
- update the README file, if necessary<br>
 *如有必要，更新自述文件*<br>
