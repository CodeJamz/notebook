## Git Tag
the `git tag` command is used to add a marker on a specific commit. The tag does not move around as new commits are added<br>
*`git tag`命令用于在特定提交上添加标记。 添加新的提交后，标记不会四处移动*
```
$ git tag -a beta
```

This command will:
- add a tag to the most recent commit (*在最新提交中添加标签*)
- add a tag to a specific commit if a SHA is passed (*如果通过了SHA，则将标签添加到特定的提交中*）

The Terminal application showing the output of the `git log` command. The log output now displays the newly created tag.<br>
*终端应用程序显示 `git log`命令的输出。 日志输出现在显示新创建的标签*

A Git tag can be deleted with the `-d` flag (for delete!) and the name of the tag:<br>
*可以使用 `-d`标志（用于删除！）和标签名称来删除Git标签*
```
$ git tag -d beta
```

## Git Branch
The `git branch` command is used to interact with Git's branches:<br>
*`git branch`命令用于与Git的分支进行交互*
```
$ git branch
```
It can be used to:
- list all branch names in the repository (*列出存储库中的所有分支名称*)
- create new branches (*创建新分支*)
- delete branches (*删除分支*)

If we type out just `git branch` it will list out the branches in a repository<br>
*如果我们只输入 `git branch`，它将列出存储库中的分支*

**Create A Branch**
**创建一个分支**<br>
To create a branch, all you have to do is use `git branch` and provide it the name of the branch you want it to create. So if you want a branch called "sidebar", you'd run this command:<br>
*要创建分支，您所要做的就是使用 `git branch`并为其提供您要创建的分支的名称。 因此，如果您想要一个名为"sidebar"的分支，则可以运行以下命令*
```
$ git branch sidebar
```

## Git Checkout
To switch between branches, we need to use Git's `checkout` command.<br>
*要在分支之间切换，我们需要使用Git的 `checkout`命令*
```
$ git checkout sidebar
```

Running this command will:
- remove all files and directories from the Working Directory that Git is tracking<br>
 *从Git跟踪的工作目录中删除所有文件和目录*
 - (files that Git tracks are stored in the repository, so nothing is lost)<br>
  (*Git跟踪的文件存储在资源库中，因此不会丢失任何内容*)
- go into the repository and pull out all of the files and directories of the commit that the branch points to<br>
 *进入存储库并拉出分支指向的所有提交文件和目录*

**So this will remove all of the files that are referenced by commits in the master branch. It will replace them with the files that are referenced by the commits in the sidebar branch.**<br>
**因此，这将删除master分支中的提交所引用的所有文件。 它将用分支中的提交引用的文件替换它们**

### Branches In The Log
### 日志中的分支
The branch information in the command prompt is helpful, but the clearest way to see it is by looking at the output of `git log`. We need it to display branches.<br>
*命令提示符中的分支信息很有用，但最清晰的方法是查看 `git log`的输出。我们需要它来显示分支*
```
$ git log
```

Notice how the special "HEAD" indicator we saw earlier has an arrow pointing to the sidebar branch. It's pointing to sidebar because the sidebar branch is the current branch, and any commits made right now will be added to the sidebar branch.<br>
*我们前面看到的特殊“ HEAD”指示器如何有一个箭头指向侧边栏分支。 它指向侧边栏，因为侧边栏分支是当前分支，并且现在进行的所有提交都将添加到侧边栏分支*

### The Active Branch
### 活动分支
The command prompt will display the active branch. But this is a special customization we made to our prompt. If you find yourself on a different computer, the fastest way to determine the active branch is to look at the output of the `git branch` command. An asterisk will appear next to the name of the active branch.<br>
*命令提示符将显示活动分支。 但这是我们对提示进行的特殊定制。 如果您发现自己在另一台计算机上，则确定活动分支的最快方法是查看`git branch`命令的输出。 活动分支名称旁边将出现一个星号*

### Delete A Branch
### 删除分支
A branch is used to do development or make a fix to the project that won't affect the project (since the changes are made on a branch).<br>
*分支用于进行开发或修复不会影响项目（因为更改是在分支上进行的）*<br>

If you want to delete the branch, you'd use the `-d` flag. The command below includes the `-d` flag which tells Git to delete the provided branch (in this case, the "sidebar" branch).<br>
*如果要删除分支，则可以使用 `-d`标志。 下面的命令包含 `-d`标志，该标志告诉Git删除提供的分支（在本例中为“ sidebar”分支）*
```
$ git branch -d sidebar
```
One thing to note is that you can't delete a branch that you're currently on. So to delete the `sidebar` branch, you'd have to switch to either the `master` branch or create and switch to a new branch.<br>
*需要注意的一件事是您无法删除当前所在的分支。 因此，要删除`sidebar`分支，您必须切换到`master`分支或创建并切换到新分支*

If you created the `sidebar` branch, added commits to it, and then tried to delete it with the `git branch -d sidebar`, Git wouldn't let you delete the branch because you can't delete a branch that you're currently on. If you switched to the `master` branch and tried to delete the `sidebar` branch, Git also wouldn't let you do that because those new commits on the `sidebar` branch would be lost! <br>
*如果创建了`sidebar`分支，向其添加了提交，然后尝试使用`git branch -d`侧边栏将其删除，则Git不允许您删除该分支，因为您无法删除当前所在的分支。 如果您切换到`master`分支并尝试删除`sidebar`分支，Git也将不允许您这样做，因为侧边栏分支上的那些新提交将丢失！*

## Merge
The `git merge` command is used to combine Git branches:<br>
*`git merge`命令用于合并Git分支*

When a merge happens, Git will:
- look at the branches that it's going to merge<br>
 *看看它要合并的分支*
- look back along the branch's history to find a single commit that both branches have in their commit history<br>
 *回顾分支的历史以查找两个分支在其提交历史中均具有的单个提交*
- combine the lines of code that were changed on the separate branches together<br>
 *将在单独的分支上更改的代码行组合在一起*
- makes a commit to record the merge<br>
 *提交记录合并的记录*
---
**Fast-forward Merge 快进合并**<br>
When a merge is performed, the other branch's changes are brought into the branch that's currently checked out.<br>
*执行合并时，另一个分支的更改将带入当前已激活的分支*<br>
We're not merging two branches into a new branch. We're not merging the current branch into the other branch.<br>
*我们不会将两个分支合并到一个新分支中。 我们没有将当前分支合并到另一个分支*<br>
Now, since 'footer' is directly ahead of 'master', this merge is one of the easiest merges to do. Merging 'footer' into 'master' will cause a Fast-forward merge. A Fast-forward merge will just move the currently checked out branch forward until it points to the same commit that the other branch (in this case, 'footer') is pointing to.<br>
*现在，由于“页脚”直接位于“母版”之前，因此此合并是最简单的合并之一。 将“页脚”合并为“主”将导致快速合并。 快速前进合并只会将当前签出的分支向前移动，直到它指向另一个分支（在此示例中为“页脚”）指向的同一提交*<br>
```
$ git merge footer
```
* **Before**<br>
(footer)<br>
(HEAD -> master)
* **After**<br>
(Head -> master, footer)<br>

The Terminal application showing the result of merging the `footer` branch _into the `master` branch.<br>
*终端应用程序显示将 `footer`合并到 `master`的结果*<br>

**Regular Merge 常规合并**<br>
To merge in the `sidebar` branch, make sure you're on the `master` branch and run:<br>
*要合并到 `sidebar`分支中，请确保您位于 `master`分支上并运行*
```
$ git merge sidebar
```

Because this combines two divergent branches, a commit is going to be made. And when a commit is made, a commit message needs to be supplied. Since this is a merge commit a default message is already supplied. You can change the message if you want, but it's common practice to use the default merge commit message. So when your code editor opens with the message, just close it again and accept that commit message.<br>
*因为这合并了两个不同的分支，所以将要进行提交。 并且，在进行提交时，需要提供提交消息。 由于这是合并提交，因此已经提供了默认消息。 您可以根据需要更改消息，但是通常的做法是使用默认的合并提交消息。 因此，当您的代码编辑器随消息一起打开时，只需再次将其关闭并接受该提交消息即可。*<br>


## Merge Conflicts
A merge conflict happens when the same line or lines have been changed on different branches that are being merged. Git will pause mid-merge telling you that there is a conflict and will tell you in what file or files the conflict occurred.<br> 
*当在合并的不同分支上更改了同一行或多行时，就会发生合并冲突。 Git将在合并过程中暂停，告诉您存在冲突，并告诉您发生冲突的文件在哪个文件中。*<br>
To resolve the conflict in a file:
- locate and remove all lines with merge conflict indicators (*找到并删除所有带有合并冲突指示符的行*)
- determine what to keep (*确定要保留什么*)
- save the file(s) (*保存文件*)
- stage the file(s) (*暂存文件*)
- make a commit (*提交*)

**Forcing A Merge Conflict 强制合并冲突**<br>
Remember that a merge conflict occurs when the exact same line(s) are changed in separate branches. Let's alter the page's heading on two different branches.<br>
*在单独的分支中更改完全相同的行时，会发生合并冲突。 让我们在两个不同的分支上更改页面的标题。*<br>
So Let's:

- change the heading on the `master` branch (*更改 `master`分支上的标题*)
- create a `heading-update` branch that's located on the commit right before the recently modified `master` branch<br>
 (*在最近修改的 `master`分支之前创建一个位于提交上的 `heading-update`分支*)
- change the same heading (*更改相同的标题*)
- switch back to the `master` branch (*切换回 `master`分支*)
- merge in the `heading-update` branch (*在 `heading-update`分支中合并*)

**Merge Conflict Indicators Explanation 合并冲突指示器说明**<br>
Git is using the merge conflict indicators to show you what lines caused the merge conflict on the two different branches as well as what the original line used to have.<br>
*Git使用合并冲突指示器来显示导致两个不同分支上发生合并冲突的行以及原始行曾经具有的行*

The editor has the following merge conflict indicators (*编辑器具有以下合并冲突指示符*):
- `<<<<<<< HEAD` everything below this line (until the next indicator) shows you what's on the current branch<br>
 *`<<<<<<< HEAD` 此行下的所有内容（直到下一个指示器）向您显示当前分支上的内容*
- `||||||| merged common ancestors` everything below this line (until the next indicator) shows you what the original lines were<br>
 *`||||||| merged common ancestors` 此行下方的所有内容（直到下一个指示器）向您显示原始行是什么*
- `=======` is the end of the original lines, everything that follows (until the next indicator) is what's on the branch that's being merged in<br>
 *`=======` 是原始行的结尾，其后的所有内容（直到下一个指示符）是要合并到的分支中的内容*
- `>>>>>>> heading-update` is the ending indicator of what's on the branch that's being merged in (in this case, the heading-update branch)<br>
 *`>>>>>>> heading-update`是合并分支中的内容的结束指示符（在本例中为heading-update分支）*

**Resolving A Merge Conflict 解决合并冲突**<br>
So to resolve a merge conflict, you need to:
- choose which line(s) to keep (*选择要保留的行*)
- remove all lines with indicators (* 删除所有带有指示器的行*)

**Commit Merge Conflict 提交合并冲突**<br>
Once you've removed all lines with merge conflict indicators and have selected what heading you want to use, just save the file, add it to the staging index, and commit it!<br>
*删除所有带有合并冲突指示符的行并选择了要使用的标题后，只需保存文件，将其添加到暂存索引中，然后提交即可！*<br>
Just like with a regular merge, this will pop open your code editor for you to supply a commit message. Just like before, it's common to use the provided merge commit message, so after the editor opens, just close it to use the provided commit message.<br>
*就像常规合并一样，这将弹出您的代码编辑器以供您提供提交消息。 与以前一样，通常使用提供的合并提交消息，因此在打开编辑器后，只需关闭它即可使用提供的提交消息。*
