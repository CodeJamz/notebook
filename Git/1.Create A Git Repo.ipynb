## Required Commands
### 所需命令
- `ls` - used to list files and directories（*用于列出文件和目录*）
- `mkdir` - used to create a new directory（*用于创建新目录*）
- `cd` - used to change directories（*用于更改目录*）
- `rm` - used to remove files and directories（*用于删除文件和目录*）

We'll also be using the idea of the current working directory, the directory that your shell is "looking at" right now. Using cd changes your working directory, and using ls (by itself) lists the files in the working directory. If you lose track of what your shell's working directory is, you can print its name with the pwd command (which stands for "print working directory").

*我们还将使用当前工作目录的概念，即您的shell当前正在“查看”的目录。使用cd更改您的工作目录，使用ls（本身）列出工作目录中的文件。如果您不了解外壳程序的工作目录，则可以使用pwd命令（其代表“打印工作目录”）打印其名称。*

## Git Init 
Use the `git init` command to create a new, empty repository in the current directory.

*使用 `git init`命令在当前目录中创建一个新的空存储库。*
```
$ git init
```

Running this command creates a hidden `.git` directory. This `.git` directory is the brain/storage center for the repository. It holds all of the configuration files and directories and is where all of the commits are stored.

*运行此命令将创建一个隐藏的 `.git`目录。 该 `.git`目录是存储库的大脑/存储中心。 它保存所有配置文件和目录，并且是所有提交的存储位置。*

## Git Clone
The `git clone` command is used to create an identical copy of an existing repository.

*`git clone`命令用于创建现有存储库的相同副本。*
```
$ git clone <path-to-repository-to-clone>
```

This command:

- takes the path to an existing repository<br>
（*采用现有存储库的路径*）
- by default will create a directory with the same name as the repository that's being cloned<br>
（*默认情况下，将创建一个与被克隆的存储库同名的目录*）
- can be given a second argument that will be used as the name of the directory<br>
（*可以给第二个参数，它将用作目录的名称*）
- will create the new repository inside of the current working directory<br>
（*将在当前工作目录内创建新的存储库*）

## Git Status
The `git status` command will display the current status of the repository. 

*`git status`命令将显示存储库的当前状态。*
```
$ git status
```
This command will:
- tell us about new files that have been created in the Working Directory that Git hasn't started tracking, yet<br>
(*告诉我们有关Git尚未开始跟踪的在工作目录中创建的新文件的信息*)
- files that Git is tracking that have been modified<br>
(*Git正在跟踪的文件已被修改*)
