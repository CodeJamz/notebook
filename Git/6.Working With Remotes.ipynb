## Ways to access a Remote

Remotes can be accessed in a couple of ways (*可以通过以下两种方式访问遥控器*):
- with a URL (*带有URL*)
- path to a file system (*文件系统的路径*)

the most common way to access a remote repository is through a URL to a repository that’s out on the web. <br>
*访问远程存储库的最常见方法是通过URL到达网络上的存储库。*<br>
The way we can interact and control a remote repository is through the `git remote` command:<br>
*我们可以通过 `git remote`命令来交互和控制远程存储库：*
```
$ git remote
```
The output of `git remote` is just a name<br>
*`git remote`的输出是名称*

If you want to see the full path to the remote repository, then all you have to do is use the `-v` flag<br>
*如果要查看远程存储库的完整路径，则只需使用`-v`标志：*
```
$ git remote -v
```
The output includes the shortname and the full URL that it refers to.<br>
*输出包括短名称和它引用的完整URL。*

## 从命令行推送现有存储库
```
git remote add origin https://github.com/CodeJamz/notebook.git
git push -u origin master
```

There are a couple of things to notice about the command you just ran on the command line:
- first, this command has the sub command `add`. (*首先，此命令具有 `add`的子命令*)
- the word `origin` is used - this is setting the shortname that we discussed earlier. (*使用 `origin`一词-这是我们前面讨论的简称)
 - Remember that the word `origin` here isn't special in any way. (请记住，这里的`origin`一词在任何方面都不是特殊的)
 - If you want to change this to repo-on-GitHub, then (before running the command) just change the word "origin" to "repo-on-GitHub": (*如果要将其更改为repo-on-GitHub，则（在运行命令之前）只需将单词“ origin”更改为“ repo-on-GitHub”*)
- third, the full path to the repository is added (i.e. the URL to the remote repository on the web). (*第三，添加到存储库的完整路径（即，网络上远程存储库的URL*）

**Sending Commits发送提交**<br>
To send local commits to a remote repository you need to use the `git push` command. You provide the remote short name and then you supply the name of the branch that contains the commits you want to push:<br>
*要将本地提交发送到远程存储库，您需要使用 `git push`命令。 您提供远程短名称，然后提供包含要推送的提交的分支名称：*
```
$ git push <remote-shortname> <branch>
```

- Depending on how you have configured GitHub and the remote URL that's being used, you might have to enter your username and password.<br>
 *根据您配置GitHub的方式和所使用的远程URL，您可能必须输入用户名和密码。*

  *如果您使用遥控器的HTTP版本（而不是ssh版本），则会发生这种情况*

One very important thing to know is that this origin/master tracking branch is not a live representation of where the branch exists on the remote repository. <br>
*要知道的一件非常重要的事情是，此源/主跟踪分支不是该分支在远程存储库中的位置的实时表示。*<br>
If a change is made to the remote repository not by us but by someone else, the origin/master tracking branch in our local repository will not move.<br>
*如果不是我们而是由其他人对远程存储库进行了更改，则本地存储库中的源/主跟踪分支将不会移动。*

## Pulling Changes From A Remote
`git push` will sync the remote repository with the local repository.<br>
*`git push`将使远程存储库与本地存储库同步。*<br>
```
$ git pull origin master
```
When `git pull` is run, the following things happen:
- the commit(s) on the remote branch are copied to the local repository (*远程分支上的提交已复制到本地存储库*)
- the local tracking branch (`origin/master`) is moved to point to the most recent commit <br>
 *本地跟踪分支（`origin/master`）被移至指向最近的提交*
- the local tracking branch (`origin/master`) is merged into the local branch (`master`) <br>
 *本地跟踪分支（`origin/master`）合并到本地分支（`master`）*

Also, changes can be manually added on GitHub (but this is not recommended). <br>
*另外，可以在GitHub上手动添加更改（但是不建议这样做）*

## Pull vs Fetch
`git fetch` is used to retrieve commits from a remote repository's branch but it does not automatically merge the local branch with the remote tracking branch after those commits have been received.<br>
*`git fetch` 用于从远程存储库的分支中检索提交，但是在收到这些提交之后，它不会自动将本地分支与远程跟踪分支合并。*<br>
You provide the exact same information to `git fetch` as you do for `git pull`. <br>
*您提供与`git pull`完全相同的信息来进行`git fetch`*<br>
```
$ git fetch origin master
```
When git fetch is run, the following things happen:
- the commit(s) on the remote branch are copied to the local repository (*远程分支上的提交已复制到本地存储库*)
- the local tracking branch (e.g. origin/master) is moved to point to the most recent commit (*本地跟踪分支（例如Origin / master）被移至指向最新提交*)

The important thing to note is that the local branch does not change at all.<br>
*要注意的重要一点是，本地分支根本不会更改*<br>
You can think of git fetch as half of a git pull. The other half of git pull is the merging aspect.<br>
*您可以将git fetch视为git pull的一半。 git pull的另一半是合并方面*

> 如果本地仓库落后于远程仓库，此时在本地仓库进行一次提交，前几个提交相同，但现在它们的`master`分支指向不同提交<br>
> 此时运行**`git pull origin master`**不会起作用<br>
> 使用`git fetch`在不合并分支的情况下获取远程仓库的提交<br>
> 所以如果我们执行 **`git fetch origin master`** 命令，就会获取远程仓库的提交，并且将 `origin/master` 指向它们<br>
> 之后运行 **`git merge origin/master`** 把远程做的修改更新到我们的 master 分支上<br>
> 现在本地仓库拥有一些远程仓库没有的提交，所以我们需要推送这些提交<br>
> 在 origin/master 分支执行 **`git push origin master`** 命令将会推送更新后的本地提交至远程仓库<br>

You can think of the git pull command as doing two things:

1. fetching remote changes (which adds the commits to the local repository and moves the tracking branch to point to them)<br>
 *获取远程更改（将提交添加到本地存储库，并将跟踪分支移至指向它们）*
2. merging the local branch with the tracking branch<br>
 *将本地分支与跟踪分支合并*

The `git fetch` command is just the first step. It just retrieves the commits and moves the tracking branch. It does not merge the local branch with the tracking branch. The same information provided to `git pull` is passed to `git fetch`:<br>
*`git fetch`命令只是第一步。 它只是检索提交并移动跟踪分支。 它不会将本地分支与跟踪分支合并。 提供给`git pull`的相同信息传递给`git fetch`*
- the shorname of the remote repository (*远程存储库的简称*)
- the branch with commits to retrieve (*具有提交检索的分支*)
