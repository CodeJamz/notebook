# 列表
## 无序列表
- 文本1
- 文本2
- 只需要在文字前面加上` - `就可以了

## 有序列表
1. 文本1
2. 文本2
- 在文字前面加上` 1. 2. 3`.


- 注：`-、1.`和文本之间要保留一个字符的空格

## 列表嵌套
- 上一级和下一级之间敲三个空格即可
   - 下一级列表
      - 下一级列表

# 链接
- 插入链接 `[显示文本](链接地址)` 

[baidu](https://www.baidu.com)

- 插入图片 `![](图片链接地址)`
![](https://www.baidu.com/img/bd_logo1.png)

# 引用
- 在你希望引用的文字前面加上 `>`

> 锄禾日当午
>> 汗滴禾下土
>>> 谁知盘中餐
>>>> 粒粒皆辛苦

# 字体
**粗体**<br>
*斜体*<br>
***斜体加粗***<br>
~删除~

# 分割线
- 三个或者三个以上的 `-` 或者 `*` <br>
---
***

# 代码
- 不分行，可以用` 将语句包起来
- 为多行，可以将```置于这段代码的首行和末行<br>
`hello world`
